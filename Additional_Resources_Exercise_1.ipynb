```python
#Create a Python function to check if a given string is a palindrome
def is_palindrome(s):
    return s == s[::-1]

# Get input from the user
string = input("Enter a string: ")

# Check and display the result
if is_palindrome(string):
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
#Calculate the compound interest for a given principal amount, interest rate, and time period
P = float(input("Principal amount: "))
R = float(input("interest rate (%): "))
T = float(input("Time period: "))
CI = P * (1 + R/100) ** T - P
print("Compound Interest:", CI)
```

```python
# Write a program that converts a given number of days into years, weeks, and days
days = int(input("Enter number of days: "))
years = days // 365
weeks = (days % 365) // 7
days = (days % 365) % 7
print("Years:", years, "Weeks:", weeks, "Days:", days)
```

```python
#Given a list of integers, find the sum of all positive numbers
numbers = [3, -1, 7, -5, 2, -8]
positive_int = 0
for num in numbers:
    if num > 0:
        positive_int += num

print(positive_int)
```

```python
# Write a program to check if a number is prime.
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")

```

```python
# Implement a program that finds the largest number in a list.
numb = [4, 17, 2, 89, 33, 21]
largest = numb[0]
for num in numb:
    if num > largest:
        largest = num

print("The largest number is:", largest)
```

```python
# Create a program that takes a year as input and checks if it is a leap year or not
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
# Calculate the sum of digits of a given number.
num = int(input("Enter a number: "))
sum_of_digits = 0
while num > 0:
    digit = num % 10
    sum_of_digits += digit
    num = num // 10

print("Sum of digits:", sum_of_digits)
```
